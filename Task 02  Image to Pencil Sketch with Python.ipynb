# Task 02 : Image to Pencil Sketch with Python

## We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

### Importing necessary libraries

```python
import cv2
from PIL import Image
```

### Reading the image

```python
image = cv2.imread("butterfly.jpg")
```

### Converting the image to RBG format

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

### Displaying the image

```python
display(Image.fromarray(image))
```

### Converting it into a grayscale image

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

#### Displaying the grayscale image

```python
display(Image.fromarray(grey_image))
```

#### Inverting the grayscale image

```python
inverted_grey_image = 255 - grey_image
```

#### Displaying the inverted grayscale image(neagtive image)

```python
display(Image.fromarray(inverted_grey_image))
```

#### Blurring the negative image

```python
blurry_negative_image = cv2.GaussianBlur(inverted_grey_image, (21, 21), 0)
```

#### Displaying the blurry negative image

```python
display(Image.fromarray(blurry_negative_image))
```

#### Inverting the blurry image

```python
inverted_blurry_image = 225 - blurry_negative_image
```

#### Displaying the inverted blurry image

```python
display(Image.fromarray(inverted_blurry_image))
```

#### Dividing the grayscale image by the inverted blurry image

```python
divided_image = cv2.divide(grey_image, inverted_blurry_image, scale = 256.0)
```

#### Displaying the divided image

```python
display(Image.fromarray(divided_image))
```

#### Comparing the original image with the pencil sketch image

#### Displaying both the images

```python
display(Image.fromarray(image))
display(Image.fromarray(divided_image))
```
